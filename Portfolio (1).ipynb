# Eksplorasi Data

```python
import pandas as pd
```

```python
dataframe = pd.read_csv('BikeSalesEurope.csv')
```

```python
dataframe.head()
```

```python
# Melihat kemungkinan data dan datatype

dataframe.info()
```

```python
# Mengubah format penanggalan

import datetime as datetime

date = []

for item in dataframe['Date']:
    item_ = item.split('-')
    year = int(item_[0])
    day = int(item_[2])
    month = int(item_[1])
    date += [datetime.date(year, month, day)]

dataframe['Date'] = date
dataframe['Date'] = pd.to_datetime(dataframe['Date'])
```

```python
# Mengubah format gender 

dataframe['Customer_Gender'] = dataframe['Customer_Gender'].replace({'M': 'Male', 'F': 'Female'})

dataframe.head()
```

```python
dataframe.info()
```

```python
# Menghapus kolom Day, Month, dan Year 

df = dataframe.drop(['Day', 'Month', 'Year'], axis = 1)
df.head()
```

```python
df.describe()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df_ = df.drop_duplicates(subset = 'Date', keep = 'last')

df_.duplicated().sum()
```

# Pemodelan

```python
data = df_[['Profit', 'Cost', 'Revenue']]
data
```

```python
plt.figure(figsize = (12, 6))

sns.heatmap(data.corr(), annot = True)
plt.show()
```

```python
from sklearn.model_selection import train_test_split as tts

X = data.drop(['Revenue'], axis = 1)
y = data['Revenue']

X_train, X_test, y_train, y_test = tts(X, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression as lr

model = lr()

model.fit(X_train, y_train)

predictions = model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

print("MAE:", mean_absolute_error(y_test, predictions))
print("MSE:", mean_squared_error(y_test, predictions))
print("R^2:", r2_score(y_test, predictions))
```

# Evaluasi Model

```python
plt.figure(figsize = (12, 6))
plt.title("Pendapatan Sebenarnya v/s Prediksi Pendapatan")

ax1 = sns.distplot(y, hist = False, color = "green", label = "Pendapatan Sebenarnya")

sns.distplot(predictions, hist = False, color = 'red', label = "Fitted Value", ax = ax1)

plt.legend(title = 'Legenda', loc = 'upper right', labels = ['Pendapatan Sebenarnya', 'Prediksi Pendapatan'])
plt.show()
```
